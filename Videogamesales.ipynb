Step-3

EDA Questions Using Matplotlib and Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
```

```python
games_df = pd.read_csv(r'E:\Videogamesales_engagement\games.csv')
```

```python
# Clean column names
```

```python
games_df.columns = games_df.columns.str.strip().str.lower().str.replace(' ', '_')
```

Questions From games.csv

1. Top-Rated Games

```python
top_rated = games_df[['title', 'rating']].dropna().sort_values(by='rating', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='rating', y='title', data=top_rated, palette='viridis')
plt.title('Top 10 Rated Games')
plt.xlabel('Rating')
plt.ylabel('Game Title')
plt.show()
```

2.Highest Average-Rated Developers

```python
dev_rating = games_df[['team', 'rating']].dropna()
avg_dev_rating = dev_rating.groupby('team')['rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_dev_rating.values, y=avg_dev_rating.index, palette='magma')
plt.title('Top 10 Developers by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Developer')
plt.show()
```

3.Most Common Genres

```python
from collections import Counter

# Split genres by comma
all_genres = games_df['genres'].dropna().str.split(', ')
flat_genres = [genre for sublist in all_genres for genre in sublist]
genre_counts = pd.Series(Counter(flat_genres)).sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='cool')
plt.title('Most Common Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

4.Backlog vs Wishlist Analysis

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=games_df, x='backlogs', y='wishlist')
plt.title('Backlogs vs Wishlist')
plt.xlabel('Backlogs')
plt.ylabel('Wishlist')
plt.show()
```

5.Release Trend Over Years

```python
games_df['release_year'] = pd.to_datetime(games_df['release_date'], errors='coerce').dt.year
release_trend = games_df['release_year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x=release_trend.index, y=release_trend.values, marker='o')
plt.title('Game Release Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Games Released')
plt.show()
```

6.Ratings Distribution

```python
plt.figure(figsize=(8, 5))
sns.histplot(games_df['rating'].dropna(), bins=20, kde=True, color='teal')
plt.title('Distribution of User Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

7. Top Wishlisted Games

```python
top_wishlist = games_df[['title', 'wishlist']].dropna().sort_values(by='wishlist', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='wishlist', y='title', data=top_wishlist, palette='cubehelix')
plt.title('Top 10 Most Wishlisted Games')
plt.xlabel('Wishlist Count')
plt.ylabel('Game Title')
plt.show()
```

8.Average Plays Per Genre

```python
genre_play_df = games_df[['genres', 'plays']].dropna()

genre_play_df['plays'] = genre_play_df['plays'].astype(str).str.replace(',', '').str.extract(r'(\d+)')
genre_play_df['plays'] = pd.to_numeric(genre_play_df['plays'], errors='coerce')

genre_play_df['genres'] = genre_play_df['genres'].str.split(', ')
genre_play_df = genre_play_df.explode('genres')

avg_plays_per_genre = genre_play_df.groupby('genres')['plays'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_plays_per_genre.values, y=avg_plays_per_genre.index, palette='plasma')
plt.title('Average Plays Per Genre')
plt.xlabel('Average Plays')
plt.ylabel('Genre')
plt.show()
```

9. Most Productive Developers (Most Games Developed)

```python
prod_dev = games_df['team'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=prod_dev.values, y=prod_dev.index, palette='Set2')
plt.title('Top 10 Most Productive Developers')
plt.xlabel('Number of Games')
plt.ylabel('Developer')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load and clean
```

```python
sales_df = pd.read_csv(r'C:\Users\JAI\Downloads\vgsales.csv')
```

```python
# Standardize columns
```

```python
sales_df.columns = sales_df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# Drop rows with missing values in key fields
```

```python
sales_df.dropna(subset=['global_sales', 'year', 'platform', 'genre'], inplace=True)
```

1. Region with Highest Sales

```python
region_sales = sales_df[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x=region_sales.values, y=region_sales.index, palette="coolwarm")
plt.title("Total Sales by Region (in millions)")
plt.xlabel("Sales")
plt.ylabel("Region")
plt.show()
```

2. Best-Selling Platforms

```python
platform_sales = sales_df.groupby('platform')['global_sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=platform_sales.values, y=platform_sales.index, palette="viridis")
plt.title("Top 10 Best-Selling Platforms Globally")
plt.xlabel("Global Sales")
plt.ylabel("Platform")
plt.show()
```

3. Sales Trends Over Time


```python
yearly_sales = sales_df.groupby('year')['global_sales'].sum().reset_index()

plt.figure(figsize=(10,5))
sns.lineplot(data=yearly_sales, x='year', y='global_sales', marker='o')
plt.title("Global Sales Trend Over Time")
plt.xlabel("Year")
plt.ylabel("Global Sales (in millions)")
plt.show()
```

4.Top Publishers

```python
top_publishers = sales_df.groupby('publisher')['global_sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_publishers.values, y=top_publishers.index, palette="mako")
plt.title("Top 10 Publishers by Global Sales")
plt.xlabel("Global Sales")
plt.ylabel("Publisher")
plt.show()
```

5. Top 10 Global Best-Sellers

```python
top_games = sales_df.sort_values(by='global_sales', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_games['global_sales'], y=top_games['name'], palette="rocket")
plt.title("Top 10 Best-Selling Games Globally")
plt.xlabel("Global Sales")
plt.ylabel("Game Title")
plt.show()
```

6. Regional Comparison by Platform

```python
platform_region = sales_df.groupby('platform')[['na_sales','eu_sales','jp_sales','other_sales']].sum()
platform_region = platform_region.loc[platform_region.sum(axis=1).sort_values(ascending=False).head(5).index]

platform_region.plot(kind='bar', stacked=True, figsize=(12,6), colormap='tab20c')
plt.title("Regional Sales Distribution by Platform")
plt.ylabel("Sales (millions)")
plt.show()
```

7. Platform Evolution Over Time

```python
platform_trend = sales_df.groupby(['year', 'platform'])['global_sales'].sum().reset_index()

top_platforms = sales_df.groupby('platform')['global_sales'].sum().sort_values(ascending=False).head(5).index
filtered = platform_trend[platform_trend['platform'].isin(top_platforms)]

plt.figure(figsize=(12,6))
sns.lineplot(data=filtered, x='year', y='global_sales', hue='platform', marker='o')
plt.title("Sales Trends of Top Platforms Over Time")
plt.xlabel("Year")
plt.ylabel("Global Sales")
plt.legend(title='Platform')
plt.show()
```

8.  Regional Genre Preferences

```python
genre_region = sales_df.groupby('genre')[['na_sales','eu_sales','jp_sales','other_sales']].sum()

genre_region.plot(kind='bar', figsize=(12,6), colormap='Set3')
plt.title("Genre Sales by Region")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

9. Yearly Regional Sales Change

```python
year_region = sales_df.groupby('year')[['na_sales','eu_sales','jp_sales','other_sales']].sum().reset_index()

plt.figure(figsize=(12,6))
for region in ['na_sales','eu_sales','jp_sales','other_sales']:
    sns.lineplot(data=year_region, x='year', y=region, label=region)

plt.title("Yearly Sales Change by Region")
plt.xlabel("Year")
plt.ylabel("Sales (in millions)")
plt.legend(title='Region')
plt.show()
```

10.Avg Sales per Publisher

```python
avg_sales_publisher = sales_df.groupby('publisher')['global_sales'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=avg_sales_publisher.values, y=avg_sales_publisher.index, palette='cubehelix')
plt.title("Top 10 Publishers by Average Global Sales")
plt.xlabel("Average Sales")
plt.ylabel("Publisher")
plt.show()
```

```python
11. Top 5 Games per Platform
```

```python
top_platforms = sales_df['platform'].value_counts().head(3).index

for platform in top_platforms:
    top_games = sales_df[sales_df['platform'] == platform].sort_values(by='global_sales', ascending=False).head(5)
    plt.figure(figsize=(8,5))
    sns.barplot(x=top_games['global_sales'], y=top_games['name'], palette='coolwarm')
    plt.title(f'Top 5 Games on {platform}')
    plt.xlabel('Global Sales')
    plt.ylabel('Game')
    plt.show()
```

5.Wishlist Correlation with Sales

```python
merged_df['wishlist'] = pd.to_numeric(merged_df['wishlist'], errors='coerce')
sns.lmplot(data=merged_df, x='wishlist', y='global_sales', height=6, aspect=1.5, scatter_kws={'alpha':0.5})
plt.title("Wishlist vs Global Sales")
plt.xlabel("Wishlist")
plt.ylabel("Global Sales")
plt.show()
```

9. Top Genre + Platform Combinations

```python
combo = merged_df.groupby(['genre', 'platform'])['global_sales'].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(12,6))
sns.barplot(data=combo, x='global_sales', y='genre', hue='platform', palette='tab20')
plt.title("Top Genre + Platform Combinations by Global Sales")
plt.xlabel("Sales")
plt.ylabel("Genre")
plt.show()
```

10.Genre-wise Regional Sales Heatmap
